```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup as bsp
import pandas as pd
```

```python
# Send Request
requests.get('https://internshala.com/internships/','html.parser')
```

```python
url='https://internshala.com/internships/'
```

# Getting Responses from the website

```python
resp=requests.get(url,'html.parser')
print(resp.status_code)
```

```python
# filtrering Field
field="web-development"
print(field)
```

# URL Handling

```python
# Create a modified URL for the specified field
modified_url=url+str(field)+'-internship/'
# Print the status of the new request
print(f"Status :{requests.get(modified_url)}")
print(modified_url)# Print the modified URL
```

```python
# Send a new request to the modified URL
resp_new=requests.get(modified_url)
```

```python
# Parse the response content using BeautifulSoup
soup=bsp(resp_new.content,'html.parser')
#print(soup)
```

```python
# type(soup)
```

```python
# Find the total number of pages available
pages=int(soup.find('span',id='total_pages').text)
# print(pages)
```

```python
# Generate a list of URLs for all pages
urlList = []
page = 1
while page <= pages:
  newUrl = modified_url+str(f"page-{page}/")
  urlList.append(newUrl)
  page +=1
# print(urlList) # Print the list of URLs
```

```python
# Initialize a list to store the parsed content of each page
soup2 = []
for url in urlList:
  resp_new=requests.get(url)
  soup3=bsp(resp_new.content,'html.parser')
  soup2.append(soup3)
# print(len(soup2)) # Print the number of pages parsed
```

```python
# print(soup.prettify())
```

# Scraping

```python
# Initialize lists to store the internship details
name=[]
for soup in soup2:
  names=soup.find_all('div',class_='individual_internship_header')
  for i in names:
    name.append(i)
# print(len(name))
#print(name)
```

```python
# print(type(name))
# profile=name.find_all('h3',class_='heading_4_5 profile')
```

```python
# for i in name:
#   p=i.find('h3',class_='heading_4_5 profile')
  # print(p)
  # print(p.text)
  # print(p.text.strip())
  # break
```

```python
# Extract and filter internship profiles
profile=[]
for i in name:
  p=i.find('h3',class_='heading_4_5 profile')
  # print(p)
  # print(p.text)
  # print(p.text.strip())
  a=p.text.strip()
  profile.append(a)
  #break
# print(len(profile))
print(f"All profiles available are : {profile}")
```

```python
# Extract company names
company=[]
for i in name:
  com=i.find('p').text.strip()
  #print(com)
  company.append(com)
# print(len(company))
print(company)
```

```python
# Extract internship details
detail=[]
for soup in soup2:
  detailList=soup.find_all('div',class_='individual_internship_internship')
  for i in detailList:
    detail.append(i)
print(len(detail))  # Print the number of internship details found
```

```python
#print(detail[0])
```

```python
# Extract locations
location=[]
for i in detail:
  loc=i.find('a').text
  location.append(loc)
  #print(loc)
# print(len(location))
print(f"Locations are : {location}")
```



> **` You have to find Duration for each individual company?`**



```python
# Extract durations of internships
duration_detail1=[]
for soup in soup2:
  duraList=soup.find_all('div',class_='item_body')
  for i in duraList:
    duration_detail1.append(i)
duration=[]
i = 1
while i < len(duration_detail1):
  duration.append(duration_detail1[i].text.strip()[0])
  i +=3
# print(len(duration))
print(duration)
```

```python
# Extract job stipends
stipend=[]
for soup in soup2:
  stipList=soup.find_all('span',class_='stipend')
  for i in stipList:
    val=i.text
    stipend.append(val)
# print(len(stipend))
print(f"Stipend is : {stipend}")
```

```python
# Extract application links
cont=[]
for soup in soup2:
  coutList=soup.find_all('div',class_='cta_container')
  for i in coutList:
    cont.append(i)
```

```python
# get Applications link
application_link=[]
for i in cont:
  anc=i.find('a')
  link=anc.get('href')
  #print(link)
  updated_link='https://internshala.com/'+link
  #print(updated_link)
  application_link.append(updated_link)
# print(len(application_link))
print(f"Application Link is : {application_link}")
```

```python
# @title Default title text
# Create a DataFrame with the extracted data
dataTable = {
    'profile': profile,
    "company": company,
    "location": location,
    "stipend":stipend,
    "duration": duration,
    "application Link": application_link
    }

df = pd.DataFrame(dataTable)
#print(f"""{len(profile)} {len(company)} {len(location)} {len(duration)} {len(stipend)} {len(application_link)}""")

print(df.head())  # Print the first few rows of the DataFrame

# Save the DataFrame to a CSV file
df.to_csv('internshala_internships.csv', index=False)
print("DataFrame saved to 'internshala_internships.csv'")

```
